```python
import pandas as pd

# Load the main file and the second file
main_df = pd.read_csv('./Radiomics_Feature.csv')
second_df = pd.read_csv('./new_labels_with_augmented.csv')

# Merge the DataFrames based on the 'Patient_id' column
filtered_df = main_df[main_df['Patient_id'].isin(second_df['Patient_id'])]

# Display the shape of the resulting dataframe (optional)
print('Filtered DataFrame shape:', filtered_df.shape)

# Save the resulting dataframe to a new CSV file
filtered_df.to_csv('Radiomics_Features_Augment.csv', index=False)

print("Filtered data has been saved to 'filtered_mainfile.csv'.")
```

```python
import pandas as pd
import shutil
import os

# Read the CSV file
csv_file = './Method-2/test_dataset_RM.csv'
df = pd.read_csv(csv_file)

# Extract the list of file names from the 'Patient_id' column
file_names = df['Patient_id'].tolist()

# Define the source and destination directories
source_dir = 'E:\my_model\My work\ROI_ny_new_files'
destination_dir = './Method-2/ROI_files'

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)

# Iterate over the list of file names and copy them to the new folder
for file_name in file_names:
    source_file = os.path.join(source_dir, file_name)
    destination_file = os.path.join(destination_dir, file_name)
    if os.path.isfile(source_file):
        shutil.copy(source_file, destination_file)
        print(f'Copied: {file_name}')
    else:
        print(f'File not found: {file_name}')

print("File copying process completed.")
```
